```python
X = df[['tbc', 'somontruot','diemrl']].values.reshape(-1,3)
Y = df['Label']
```

```python
# model = LinearRegression()
# xtrain, xtest, ytrain, ytest = train_test_split(X,Y,test_size=0.3, random_state=0)
# model.fit(xtrain, ytrain)##xay dung mo hinh hoi quy
```

```python
clf = Perceptron() 
xtrain, xtest, ytrain, ytest = train_test_split(X,Y,test_size=0.2, random_state=0)
clf.fit(xtrain, ytrain)
```

```python
ypred = clf.predict(xtest)
```

```python
from sklearn.metrics import classification_report
print(ytest.shape)
print(ypred.shape)
print(classification_report(ytest, ypred))
```

```python
master= tk.Tk()
master.title("Bài tập lớn")
master.geometry("600x400",)
tk.Label(master, text="Nhập các thông tin để dự đoán: ").grid(row=0, column=0)
tk.Label(master, text="Trung bình cộng").grid(row=1, column=0)
tk.Label(master, text="Số môn trượt").grid(row=2, column=0)
tk.Label(master, text="Điểm rèn luyện").grid(row=3, column=0)
tk.Label(master, text="Kết quả dự đoán : ").grid(row=5, column=0)
e1 = tk.Entry(master)
e2 = tk.Entry(master)
e3 = tk.Entry(master)
e5 = tk.Entry(master,state='disabled')

e1.grid(row=1, column=1,pady=(13,5))
e2.grid(row=2, column=1,pady=(0,5))
e3.grid(row=3, column=1,pady=(0,5))
e5.grid(row=5, column=1,pady=(0,5))
tk.Label(master, text="Chọn chức năng :").grid(row=6,column=0)


def duDoan():
    check1=e1.get()
    check2=e2.get()
    check3=e3.get()
    check5=e5.get()
    if(check1 == "" or check2 == "" or check3== "" or float(check1) <=0 or float(check2) <=0 or float(check3) <=0):
        messagebox.showerror("Error", "Vui lòng nhập đúng số ")
    else:
        bien1 = float(check1)
        bien2 = float(check2)
        bien3 = float(check3)
        kqDuDoan = clf.predict([[bien1,bien2,bien3]])
        e5.configure(state=tk.NORMAL)
        if(check5 !=" "):
            e5.delete(0,'end')
            e5.insert(0,round(float(kqDuDoan),1))
        else:
            e5.insert(0,round(float(kqDuDoan),1))
def Close():
    master.destroy()

def reset():
    check1=e1.get()
    check2=e2.get()
    check3=e3.get()
    check5=e5.get()
    if(check1 == "" and check2 == "" and check3== "" and check5== ""):
        messagebox.showinfo("reset","Các ô đã được reset")
    else:
        e1.delete(0,'end')
        e2.delete(0,'end')
        e3.delete(0,'end')
        e5.delete(0,'end')
        e5.configure(state=tk.DISABLED)
        

tk.Button(master, 
          text='Dự đoán', 
          command=duDoan).place(x=150,y=175)         
tk.Button(master, 
          text='Reset', 
          command=reset).place(x=240,y=175)   
tk.Button(master, 
          text='Thoát', 
          command=Close).place(x=310,y=175)

master.mainloop()
```
